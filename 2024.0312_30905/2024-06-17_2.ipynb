## 산점도 그리기

```python
import matplotlib.pyplot as plt
plt.rc('font',family = 'Malgun Gothic')
# 랜덤 라이브러리 불러오기
import random
# 데이터 만들기
height = []
weight = []
for i in range(100) :
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
# 키와 몸무게의 상관 관계
plt.title('키와 몸무게의 상관관계')
# 산점도 그래프 그리기    
plt.scatter(height,weight, c= height, cmap='RdPu')
# 축 제목
plt.colorbar(label='키')
plt.xlabel('키')
plt.ylabel('몸무게')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font',family = 'Malgun Gothic')
# 랜덤 라이브러리 불러오기
import random
# 데이터 만들기
height, weight, score = [],[],[]
for i in range(100) :
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    score.append(random.randint(1,1000))
# 키와 몸무게의 상관 관계
plt.title('키와 몸무게의 상관관계')
# 산점도 그래프 그리기    
plt.scatter(height,weight, s=score, c=score, cmap='rainbow',alpha=0.7)
# 축 제목
plt.colorbar(label='체육 실기 점수')
plt.show()
```

## 그래프 겹쳐 그리기
#### 꺽은선 그래프 겹쳐 그리기

```python
import matplotlib.pyplot as plt
singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]

plt.plot(singer, week1)
plt.plot(singer, week2)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rc('font',family = 'Malgun Gothic')
singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]

plt.title('오디션 프로그램 득표 현황')

plt.plot(singer, week1,label='첫째주',color = 'hotpink')
plt.plot(singer, week2,label='둘째주',color = 'Royalblue')
plt.legend()
plt.show()
```
